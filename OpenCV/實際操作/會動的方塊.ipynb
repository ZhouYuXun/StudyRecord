```python
import cv2
import numpy as np
import IPython.display as dp
import time

from google.colab.patches import cv2_imshow

while True:

  img = np.full((300, 300, 3), 128, np.uint8)

  for x in range(int(img.shape[1]/3*2)):

    img = np.full((300, 300, 3), 255, np.uint8)

    start = (x, int(img.shape[1]/3))
    end = (int(img.shape[1]/3)+x, int(img.shape[1]/3*2))
    color = (255, 0, 0)

    img = cv2.rectangle(img, start, end, color, -1)

    cv2_imshow(img) 

    dp.clear_output(True)

    time.sleep(0.03)

    if end[0] == 299:

      for y in range(200):

        img = np.full((300, 300, 3), 255, np.uint8)

        start = (199-y, int(img.shape[1]/3))
        end = (299-y, int(img.shape[1]/3*2))
        color = (255, 0, 0)

        img = cv2.rectangle(img, start, end, color, -1)

        cv2_imshow(img) 

        dp.clear_output(True)

        time.sleep(0.03)
```
