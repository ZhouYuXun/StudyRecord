```python
import cv2
import pytesseract as pt

from google.colab.patches import cv2_imshow

from pyzbar import pyzbar
```

```python
# 第三方庫文字辨識

img = cv2.imread('1.jpg', 1)

text = pt.image_to_string(img, "eng")

print('辨識結果', text)

cv2_imshow(img)
```

```python
# 第三方庫QECODE辨識 (可連接攝像頭) cv2.VideoCapture(0)為讀取本機鏡頭畫面，須配合cv2.waitKey(100)

img = cv2.imread('2.png', 1)

decoded_objects = pyzbar.decode(img)

for x in decoded_objects:

  print('條碼類型：', x.type)
  print('資料：', x.data.decode('utf-8'))

  x, y, w, h = x.rect

  cv2.rectangle(img, (x,y), (x+w, y+h), (0, 0, 255), 5)

cv2_imshow(img)
```

```python
# 官方特徵分類器辨識

img = cv2.imread('3.jpg', 1)

face_cascade = cv2.CascadeClassifier('/content/haarcascades/haarcascades/haarcascade_smile.xml')

ret = face_cascade.detectMultiScale(img, minNeighbors=5, minSize=(1,1))

for x,y,w,h in ret:

  cv2.rectangle(img, (x,y), (x+w, y+h), (0, 0, 255), 5)

cv2_imshow(img)
```
