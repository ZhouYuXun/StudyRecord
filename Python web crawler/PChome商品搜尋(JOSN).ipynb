```python
import requests, prettytable

find_commodity = input('關鍵字:')

page = '1'

while page != '0':

  url = 'https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=' + find_commodity + '&page='+ page +'&sort=sale/dc'
  res = requests.get(url.format(1))

  try:
    res_josn = res.json()
  except:
    print('查無此頁碼')
    break

  pt = prettytable.PrettyTable(['名稱', '價格'], encoding='UTF-8')

  for prod in res_josn['prods']:
    data = [prod['name'], prod['price']]
    pt.add_row(data)

  print(pt)

  page = input('前往頁碼:')
```
