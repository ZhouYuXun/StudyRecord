```python
import requests, os, zipfile
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import cv2

potted_flowers = ['雅樂之舞', '十二之卷', '月兔耳']

for potted in potted_flowers:
  potted, n = '多肉植物 ' + potted, 0
  

  for num in range(1, 2):
    print('{}下載第{}頁中'.format(potted, num))

    # 獲取HTML資料並使用BeautifulSoup
    url = 'https://www.bing.com/images/async?q={}&first={}&count=35&cw=1080&ch=720&relp=35&apc=2&datsrc=I&layout=RowBased&mmasync=1&dgState=x*0_y*0_h*0_c*0_i*0_r*0&IG=880ADFBA2C9A486A8634B98CEE824D0B&SFX={}&iid=images.5574'.format(potted, 35*num-35, num)
    print(url, '\n')

    # url = 'https://www.bing.com/images/search?q={}&form=HDRSC1&first={}'.format(potted, num)
    response = requests.get(url)
    html = BeautifulSoup(response.text)

    # 確認資料夾是否存，如不存在則創建
    folder_name = '多肉植物/' + potted
    if not os.path.exists(folder_name):
      os.makedirs(folder_name)

    # 本頁所有圖片超連結區塊
    a = html.find_all('a', class_ = 'iusc')

    print(len(a), '\n')

    for m in a:
      # 取得圖片網址
      image_url = eval(m.get('m'))['murl']
      print('圖片網址：', image_url)

      # 自動命名
      parsed_url = urlparse(image_url)
      file_name = os.path.basename(parsed_url.path)
      print('圖片名稱：', file_name)

      n += 1

      # 檢查檔案名稱是否為空值
      if file_name != b'':
        file_name = file_name.replace(".jpg", ".png")
        try:
          response = requests.get(image_url)
        except:
          print("網址異常")
        # 資料夾與檔案名稱
        image_path = os.path.join(folder_name, file_name)
        print('資料夾名稱', image_path)

        # 檢查檔案名稱是否存在
        if os.path.exists(image_path) != True:
          with open(image_path, "wb") as f:
            f.write(response.content)
        else:
          old_image = cv2.imread(image_path)
          
          if old_image != response.content:
            file_name = file_name.replace(".jpg", str(n)+".png")
            print('新的檔名' ,file_name)
            image_path = os.path.join(folder_name, file_name)
            try:
              with open(image_path, "wb") as f:
                f.write(response.content)
            except:
              print("檔名為空")
          else:
            print('圖片相同')
      else:
        print('檔名為空')
      print('\n')
```
