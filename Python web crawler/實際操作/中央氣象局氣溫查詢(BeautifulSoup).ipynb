```python
import requests, prettytable

from bs4 import BeautifulSoup

url = 'https://www.cwb.gov.tw/V8/C/W/TemperatureTop/County_TMax_T.html?ID=Mon'

respond = requests.get(url)

html = BeautifulSoup(respond.text, 'html.parser')

pt = prettytable.PrettyTable(['地區', '氣溫'], encoding = "UTF-8")

for tr in html.find_all('tr'):

  th = tr.find('th').text
  sp = tr.find('span').text

  data = [th, sp]

  pt.add_row(data)

print(pt)
```
