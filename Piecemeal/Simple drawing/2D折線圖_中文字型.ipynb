```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import matplotlib
```

```python
# 下載台北黑體字型
!wget -O taipei_sans_tc_beta.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download

# 新增字體
matplotlib.font_manager.fontManager.addfont('taipei_sans_tc_beta.ttf')
matplotlib.rc('font', family='Taipei Sans TC Beta')
```

```python
x = np.arange(50, 201, 1)
y = 10000/x
data = pd.DataFrame(pd.DataFrame([x, y]).T, columns=['x', 'y'])

plt.figure(figsize=(12, 6)) # 設置圖片大小

sns.lineplot(x=x, y=y, data=data)

plt.xlabel('速度')
plt.ylabel('進度條')
plt.xticks(np.arange(50, 210, 10)) # 設置x軸刻度
plt.yticks(np.arange(50, 210, 10)) # 設置y軸刻度
plt.grid(True, linestyle='--', axis='y') # 增加水平方向的網格線
plt.grid(True, linestyle='--', axis='x') # 增加垂直方向的網格線

plt.show()
```
