```python
# 匯入函數

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
column = ['time','longitude','latitude','scale','depth']

CWBdata = pd.read_csv('D:/Dataset/Earthquake/Original/CWB_593634_19990101_20201231.csv', encoding='big5')
Sortdata = pd.DataFrame(np.array(CWBdata[:len(CWBdata)].iloc[::-1]), columns=column)

frequencytotal = np.unique(Sortdata['scale'], return_counts=True)
frequencydict = dict(zip(frequencytotal[1], frequencytotal[0]))

filterscale = Sortdata['scale'] >= frequencydict.get(max(frequencytotal[1]))
filterdata = pd.DataFrame(np.array(Sortdata[filterscale]), columns=column).dropna(axis = 0)
```

```python
Nact = pd.read_excel('C:/Users/btea4/OneDrive/桌面/GRtotal2.xlsx')['cumsum']

scale = pd.read_excel('C:/Users/btea4/OneDrive/桌面/GRtotal2.xlsx')['scale']

Npre = []

for M in scale:

    N = pow(10,7.202211152358169-0.9216557919629628*M)
    
    Npre.append(N)

plt.figure(figsize = (20, 10))

plt.rcParams['font.sans-serif'] = 'DFKai-SB'

plt.xlabel("規模",fontsize=32,labelpad=20)
plt.ylabel('\n'.join(('累','積','次','數')),fontsize=32,labelpad=40,rotation='horizontal')

plt.tick_params(direction='in',length=10,labelsize=24,color='k',pad=20)

plt.plot(scale,Npre)
plt.plot(scale,Nact)

#plt.axis([2,4,-5000,200000])

plt.show()
```
